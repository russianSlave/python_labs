# Метод Сопряженных градиентов. 




Метод сопряженных градиентов  - это итерационный численный метод (первого порядка) решения оптимизационных задач, который позволяет определить экстремум (минимум или максимум) целевой функции: 
$$f(х_{1}^*,x_{2}^*,.....x_{n}^*)->extr.\; $$
В соответствии с рассматриваемым методом экстремум (максимум или минимум) целевой функции определяют в направлении наиболее быстрого возрастания (убывания) функции, т.е. в направлении градиента (антиградиента) функции. Градиентом функции  в точке 
называется вектор, проекциями которого на координатные оси являются частные производные функции по координатам: 
$$∇f(x_{1},x_{2},....x_{n})=(∂f/∂x_{1}*i+∂f/∂x_{2}*j+.....∂f/∂x_{n}*n)/.\;  $$

где i, j,…, n - единичные векторы, параллельные координатным осям.
Градиент в базовой точке  $$X=(x_{1}^*,x_{2}^*,...x_{n}^*)  \;$$ строго ортогонален к поверхности, а его направление показывает направление наискорейшего возрастания функции, а противоположное направление (антиградиент), соответственно, показывает направление наискорейшего убывания функции. 

Метод сопряженных градиентов является дальнейшим развитием метода наискорейшего спуска, который сочетает в себе два понятия: градиент целевой функции и сопряженное направление векторов. В общем случае процесс нахождения минимума функции является итерационной процедурой, которая записывается в векторной форме следующим образом: 

$$ X_{k+1}=X_{k}±λ_{k}P_{k} .\; $$
где знак «+» используется для поиска максимума функции, а знак «-» используется для поиска минимума функции. 
$$P_{k}$$- единичный вектор сопряженных направлений, который определяется по формуле: 
$$P_{0}=∇f(X_{0})\;$$
$$P_{k}=∇f(X_{k})+B_{k}P{k-1} k=1,2,3...\;$$
Существует несколько способов определения значений весовых коэффициентов (переменная ), которые используются для определения сопряженного направления.
В данной работе используется определение весового коэффициента по формуле Флетчера-Ривса 
$$B_{k}=∥∇f(x_{1},x_{2},...x_{n})_{k}∥^2 / ∥∇f(x_{1},x_{2},...x_{n})_{k-1} ∥^2 $$
модуль градиента определяет скорость возрастания или убывания функции в направлении градиента или антиградиента соответственно. 
В соответствии с представленными  выражениями новое сопряженное направление получается сложением градиента (антиградиента) в точке поворота и предыдущего направления движения, умноженного на коэффициент. Таким образом, метод сопряженных градиентов формирует направление поиска к оптимальному значению используя информацию о поиске полученную на предыдущих этапах спуска. 

Отдельной задачей являеться нахождение λ, данная задача являеться задачей одномерной минимизации.В данной работе для одномерной минимизации используется метод квадратичной ин-терполяции.
Интерполяция функции Y(x) одной переменной х, заданной (n+1) узлами y(x), где i=0,1,2,…n, заключается в нахождении значений Y по значениям х, находящихся в отрезке [X0,..Xn). При интерполяции функция f(x) заменяется интерполяционным полиномом p(x),значения которого p(x) в узлах точно совпадают с y(x). Значение n задает степень полинома p(x). Существует ряд специальных видов полинома p(x).

Метод заключается в замене f(x) в промежутке X1+(-)h, где X1 — начальное приближение, квадратичной параболой, экстремум которой вычисляется аналитически. После приближенного нахождения экстремума Xm можно задать X1=Xm и повторить поиск. Таким образом, с помощью итерационной процедуры значение Xm уточняется до получения его с заданной погрешностью e.

Этот метод обеспечивает поиск как максимумов, так и минимумов f(x), в том числе для случая f(x)=0, причем точка Xm может лежать в интервале X1+(-)h (интерполяция) и быть вне его (экстраполяция).


Для нахождения интервала унимодальности необходимо использовать аллгоритм Свенна.
  Исходные данные. х0 - начальная точка, h - шаг поиска (h>0).
  
Шаг 1. Вычислить f(x0); f(x0+h); f(x0-h); k=1.

Шаг 2. Если  f(x0-h) <=  f(x0) <=  f(x0+h), то x1 = x0 + h, перейти к шагу 4.

Шаг 3. Если f(x0-h)  >= f(x0) >= f(x0+h), то  х1 = x0-h, h = -h, перейти к шагу 4, в противном случае  (f(x0-h)   f(x0)   
f(x0+h))  a = x0 – h; b = x0 + h, конец.

Шаг 4. xk+1 = xk + 2kh , вычислить f(xk+1).

Шаг 5. Если f(xk+1)>=  f(xk), то к = к + 1, перейти к шагу 4.

Шаг 6. Если h > 0, то a = xk-1 , b = xk+1 , конец, в противном случае a = xk+1 , b = xk-1 , конец.
         Заметим, что случай f(x0-h) >= f(x0)<=  f(x0+h)
(шаг 3) не рассматривается, так как он противоречит предположению об унимодальности функции f(x).
 
Квадратичная интерполяция 
В методе задаются 3 пробные точки $\mathbf{x}_1=(a+b)/2$,$\mathbf{x}_2$ и $\mathbf{x}_3$. Для нахождения точки $\mathbf{x}_2$ задается шаг  h>0 в положительном направлении от точки $\mathbf{x}_1$, т.е. $\mathbf{x}_2=\mathbf{x}_2+h$ и если f($\mathbf{x}_1$)>f($\mathbf{x}_2$),то $\mathbf{x}_3=\mathbf{x}_1+2h$, иначе $\mathbf{x}_3=\mathbf{x}_1-h$.

Вычисляются значения функции в этих точках f($\mathbf{x}_1$),f($\mathbf{x}_2$),f($\mathbf{x}_3$), строится квадратичный интерполяционный многочлен по трем точкам и находится его точка минимума по формуле 

$\mathbf{x}^*$=$(\mathbf{x}_2^2-\mathbf{x}_3^2)f(\mathbf{x}_1)+(\mathbf{x}_3^2-\mathbf{x}_1^2)f(\mathbf{x}_2)+(\mathbf{x}_1^2-\mathbf{x}_2^2)f(\mathbf{x}_3) \over 2((\mathbf{x}_2-\mathbf{x}_3)f(\mathbf{x}_1)+(\mathbf{x}_3-\mathbf{x}_1)*f(\mathbf{x}_2)+(\mathbf{x}_1-\mathbf{x}_2)f(\mathbf{x}_3))$.

Находится также точка $\mathbf{x}_{min}=argmin(f(\mathbf{x}_1),f(\mathbf{x}_2),f(\mathbf{x}_3))$.

Если знаменатель в формуле для нахождения минимума квадратичного интерполяционного многочлена равен нулю, т.е. все три точки лежат на одной прямой. Рекомендуется выбрать за $\mathbf{x}_1=\mathbf{x}_{min}$ и повторить нахождение точки $\mathbf{x}^*$.

Критерием окончания этого процесса является выполнение условия для заданного $e$

$|\mathbf{x}_{min}-\mathbf{x}^*|<e$.

```python
from decimal import Decimal
import numpy as np
from numpy import linalg
import pylab
import math
import matplotlib.pyplot as plt
from matplotlib import cm
```

#### 2. Функция и ее частные производные, модуль градиента

```python
points = [-1,1,3]



def func(x1,x2,x3):
    x, y, z =  x1,x2,x3
    return -2*(x**2) + 16*x - (y**2) + 6*y - 3*(z**4) - 48 *(z**3) - 288*(z**2) - 768*z - 805


def dX1dF(x1,x2,x3):
    return -4 * x1 + 16

def dX2dF(x1,x2,x3):
    return -2*x2+6
def dX3dF(x1,x2,x3):
    return -12*(x3 ** 3) - 144 * (x3 **2) - 576 *x3 -768
```

```python
DfuncS = [dX1dF,dX2dF,dX3dF]

def gradAbs (x, y, z):
    return (dX1dF(x,y,z)**2+dX2dF(x,y,z)**2+dX3dF(x,y,z)**2)**0.5
```

#### 3. Нахождение интервала унимодальности методом Свенна 

```python

def SvennMethod (argNum, startPoints):
    h=0.0001
    
    params1 = [*startPoints]
    params2 = [*startPoints]  
    
    def reloadParams():
        params1 = [*startPoints]
        params2 = [*startPoints]       
    
    
    
    paramMinusH = [*startPoints] 
    paramPlusH = [*startPoints] 
    
    #x0-h
    paramMinusH[argNum] = paramMinusH[argNum] - h
    #x0+h
    paramPlusH[argNum] = paramPlusH[argNum] + h  
    

    f0,fplus,fminus = func(*startPoints),func(*paramPlusH),func(*paramMinusH)
    if (fminus<=f0) and (f0<=fplus):
        #print("svenn method h = ",h)
        print()
    else:
        if (fminus>=f0) and (f0>=fplus):
            h = -h
            #print("svenn method h = ",h)
        else:
            if (fminus<=f0) and (f0>=fplus):
                #print("emergensy exit")
                print("Интервал унимодальности по x",argNum,":[",(startPoints[argNum]-h),";",(startPoints[argNum]+h),"]")
                return (startPoints[argNum]-h),(startPoints[argNum]+h) 
            else:
                if (fminus>=f0) and (f0<=fplus):
                    print("Противоречит предположению об унимодальности")
                    return False                 
    
    xArray = []
    ##x0
    xArray.append(startPoints[argNum])
    ##x1
    xArray.append(startPoints[argNum]+h)  
    
    iteration = 1
    
    state = True
    while state:
      
        XKplus1 = xArray[len(xArray)-1] + float((2**iteration))*h
        XKplus1 = round(XKplus1,4)

        xArray.append(XKplus1)
        
        reloadParams()     
        ##new
        params1[argNum] = XKplus1
        ##old
        params2[argNum] = xArray[len(xArray)-2]
        
        iteration = iteration + 1
        
        if func(*params1)<func(*params2):
            #print("old:  ",func(*params1),"new: ",func(*params2))
            state = False
        #if iteration==10:
        #    break            
    
    #leftBorder, rightBorder = xArray[len(xArray)-3], xArray[len(xArray)-1]
    #if h < 0:
    #     leftBorder, rightBorder = xArray[len(xArray)-1], xArray[len(xArray)-3]
    
    leftBorder=min(xArray[len(xArray)-3],xArray[len(xArray)-1])
    rightBorder=max(xArray[len(xArray)-3],xArray[len(xArray)-1])
        
    print("Интервал унимодальности по x" +str(argNum)+ ":[",leftBorder,";",rightBorder,"]  итераций ",iteration)

    return leftBorder, rightBorder
    

```

#### 4. Нахождение λx и λy  методом квадратичной интерполяции

```python
def quadInterp (argNum,startPoints,leftBorder, rightBorder, pk):
    xx = 0
    eps = 0.01
    h=0.0001


    point1=(leftBorder+rightBorder)/2.0
    point2=point1+h
    point3 = 0
    
    f1args, f2args = [*startPoints],[*startPoints]
    f1args[argNum], f2args[argNum] = point1, point2

    if func(*f1args)>func(*f2args):
        point3 = point1 + 2*h
    else:
        point3 = point1 - h       

    
        
    f1args, f2args,f3args = [*startPoints],[*startPoints],[*startPoints]
    f1args[argNum] = point1
    f2args[argNum] = point2
    f3args[argNum] = point3

    
    f1=func(*f1args)
    f2=func(*f2args)
    f3=func(*f3args)
    
    
    a=[point1,point2,point3]
    xmin=a[np.argmin([f1,f2,f3])]
    
    ###polinom
    chisl = (point2**2-point3**2)*f1+(point3**2-point1**2)*f2+(point1**2-point2**2)*f3
    znam = 2.0*((point2-point3)*f1+(point3-point1)*f2+(point1-point2)*f3)
    
    
    if znam==0:
        point1=xmin
    else:
        xx=chisl/znam
     
    iter = 1
    while abs(xmin-xx)>eps:
        point2=point1+h

        f1args, f2args = [*startPoints],[*startPoints]
        f1args[argNum], f2args[argNum] = point1, point2

        if func(*f1args)>func(*f2args):
            point3 = point1 + 2*h
        else:
            point3 = point1 - h       



        f1args, f2args,f3args = [*startPoints],[*startPoints],[*startPoints]
        f1args[argNum], f2args[argNum],f3args[argNum] = point1, point2, point3
        
        f1=func(*f1args)
        f2=func(*f2args)
        f3=func(*f3args)
        

        a=[point1,point2,point3]
        xmin=a[np.argmin([f1,f2,f3])]

        ###polinom
        chisl = (point2**2-point3**2)*f1+(point3**2-point1**2)*f2+(point1**2-point2**2)*f3
        znam = 2.0*((point2-point3)*f1+(point3-point1)*f2+(point1-point2)*f3)

        iter = iter +1
        
        if znam==0:
            point1=xmin
            continue
        else:
            xx=chisl/znam
        point1=(xmin+xx)/2.0
        
    print("Максимум по x"+str(argNum)+": x"+str(argNum)+" =",xx," итераций ",iter)
    return float((min(xx,xmin)-startPoints[argNum])/DfuncS[argNum](*startPoints))
```

#### 5. Метод сопряженных градиентов, формула Флетчера-Ривса

```python
def methodSoprGrad(startPoints):
    eps=0.01
    iter=0 
    
    # x,y,z
    Xk = [*startPoints]
    Lambda = [0,0,0]
    
    ##p0
    p = [dX1dF(*startPoints),dX2dF(*startPoints),dX3dF(*startPoints)]
    
    #[x],[y],[z] storage
    storage = [[],[],[]]

    #x    
    storage[0].append(Xk[0])
    #y
    storage[1].append(Xk[1])
    #z
    storage[2].append(Xk[2])    
    
    print("X0=",Xk[0],"  X1=",Xk[1],"  X2=",Xk[2],"  func=",func(Xk[0],Xk[1],Xk[2]),"  grad=",gradAbs(*Xk),"  iter=",iter) 
    
    while(gradAbs(*Xk)>eps ):
        print("grad ",gradAbs(*Xk)-eps)
        iter+=1
        
        ##lambda
        L0,R0 = SvennMethod (0, [*Xk])
        L1,R1 = SvennMethod (1, [*Xk]) 
        L2,R2 = SvennMethod (2, [*Xk])
        Lambda = [quadInterp (0,[*Xk], L0,R0,p), quadInterp (1,[*Xk],L1,R1,p), quadInterp (2,[*Xk],L2,R2,p)]
        
        newXk = [0,0,0]
        newXk[0] = Xk[0]+Lambda[0]*p[0]
        newXk[1] = Xk[1]+Lambda[1]*p[1]
        newXk[2] = Xk[2]+Lambda[2]*p[2]
        
        ##сброс n+1 шаг
        if(iter%3==0):
            beta=0
        else:
            beta = (dX1dF(*newXk) + dX2dF(*newXk) + dX3dF(*newXk))**2
            beta/= (dX1dF(*Xk) + dX2dF(*Xk) + dX3dF(*Xk))**2
            
        ##pk    
        p[0]=dX1dF(*newXk)+beta*p[0]
        p[1]=dX2dF(*newXk)+beta*p[1]
        p[2]=dX3dF(*newXk)+beta*p[2]
        
        #update xk
        Xk = [*newXk]
     

        
        storage[0].append(Xk[0])
        storage[1].append(Xk[1])
        storage[2].append(Xk[2])
        
        print("X0=",Xk[0],"  X1=",Xk[1],"  X2=",Xk[2],"  func=",func(*Xk),"  grad=",gradAbs(*Xk),"  iter=",iter) 
    return storage
```

#### 6.  Построение линий уровня и траектории поиска

```python
dataForGraphics = methodSoprGrad(points)



def buildGraph (indFirst,indSecond,data):
    param1=np.linspace(-10,10)
    param2=np.linspace(-10,10)
    param3=np.linspace(-10,10)
    level=np.zeros((len(param1),len(param2)),dtype=float)
    
    for a in range(len(param1)):
        for b in range(len(param2)):
             for c in range(len(param3)):       
                level[a,b]=func(param1[a],param2[b],param3[c])
    fig, ax = plt.subplots()
    CS = ax.contour(param1, param2, level,10)
    ax.clabel(CS, inline=1, fontsize=10)
    ax.set_title('Линии уровня и траектория поиска при x'+str(indFirst)+'='+str(points[indFirst])+' и  при x'+str(indSecond)+'='+str(points[indSecond]))
    plt.plot(data[indSecond], data[indFirst], 'C3', zorder=2, lw=1)
    plt.scatter(data[indSecond], data[indFirst], s=20, zorder=10)
    plt.xlabel('x'+str(indSecond))
    plt.ylabel('x'+str(indFirst))
    plt.show()

        

buildGraph (0,1,dataForGraphics)
buildGraph (0,2,dataForGraphics)
buildGraph (1,2,dataForGraphics)
```
