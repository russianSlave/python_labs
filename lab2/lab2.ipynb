```python
import matplotlib.pyplot as plt
import math

def func(x):
    return (3*x**4)+(4*x**3)-(12*x**2)-7

def DDfunc(x):
    return (36*x**2)+(24*x)-24

e = 0.0001

a = -1
b = 1

while(func(a)*func(b)>0):
    a+=1
    b+=1

print ("interval ["+str(a)+";"+str(b)+"]")     

def methodHord(left, right, eps):
    static = None
    xi = None
    LastXi =None
    cycleBool = True
    state = 1
    if func(left)*DDfunc(left)>0:
        xi = left
        static = right
        state = 1
    else:
        xi = right
        static = left
        state = 2
    i = 0
    while cycleBool:
        LastXi = xi
        if state == 1:
            xi = xi - (func(xi)/(func(static)-func(xi)))*(static-xi)
        if state == 2:
            xi = xi - (func(xi)/(func(xi)-func(static)))*(xi-static)
        i +=1
        if abs(xi-LastXi)<eps:
            cycleBool = False   
    print ("method hord ITERATIONS:"  + str(i))   
    return xi

def methodDihotomia(left, right, eps):
    state = True
    i = 0
    x = None
    while state: 
        x = (left + right) / 2;
        f= func(x)
        if (func(left)*f<0):
            right = x
        else:
            left = x
        i+= 1
        if abs(right-left)<e:
            state = False
    print ("method dihotomia ITERATIONS:"  + str(i))    
    return x
    

    
korenDihot = methodDihotomia(a, b, e)   
korenHord = methodHord(a, b, e)
fXdihot = func(korenDihot)
fXhord = func(korenHord)
print ("cherez method dihotomii x: "+ str(korenDihot))
print ("cherez method dihotomii f(x): "+ str(fXdihot))
print ("cherez method hord x: "+ str(korenHord))
print ("cherez method hord f(x): "+ str(fXhord))
##################################################################
##################################################################
a = None
b = None
# 5^x-(6x+7)=0
def funcZadanie2(x):
    return (5**x)-6*x


def func2(x):
    return (5**x)-6*x-7

def func3(x):
    return (7-(5**x))/(-6)

def Dfunc2(x):
    return (5**x)*math.log(5) - 6



xArray = []
yArray = []
#OX
y1Array = []

p = -2
while p<1:
    xArray.append(p)
    y1Array.append(0)
    yArray.append(func2(p))

    p+=0.1
    
plt.plot (xArray,yArray)
plt.plot (xArray,y1Array)

plt.show()


print("Vvedi interval")

a = int(input())
b = int(input())

print ("interval ["+str(a)+";"+str(b)+"]")   
   
def methodIterations (left,eps):
    cycleBool = True
    last = None
    x = left
    iters=0
    while cycleBool:
        last = x
        x = func3(last)
        iters+=1
        if abs(last - x) < eps:
           cycleBool = False
    print ("method prostih iteracii ITERATIONS:"  + str(iters))   
    
    return x

def methodNewton (left,eps):
    cycleBool = True
    iters = 0
    x = left
    last = None
    while cycleBool:
        last = x
        x = last - (func2(last)/Dfunc2(last))
        iters+=1
        if abs(x-last) < eps:
           cycleBool = False
    print ("method Newtona  ITERATIONS:"  + str(iters))   
    return x

korenMethodIter = methodIterations(a,e)
korenMethodNewt = methodNewton(a,e)
print ("method iteracii x:" + str(korenMethodIter))
print ("method iteracii f(x):" + str(funcZadanie2(korenMethodIter)))
print ("method newtona x:" + str(korenMethodNewt))
print ("method newtona f(x):" + str(funcZadanie2(korenMethodNewt)))

########################

def fZadanie3(x,y):
    return x**4+y**4

def gZadanie3(x):
    return x**4+y**4

def gRightZadanie3(x):
    return (math.e)**(-x)

def f(x,y):
    return x**4+y**4-5

def g(x,y):
    #return y-(math.e**(-x))
    return y-(math.e**(-x))

def dxf(x,y):
    return 4*x**3

def dyf(x,y):
    return 4*y**3

def dxg(x,y):
    return math.e**(-x)

def dyg():
    return 1
            
def obratnaiaMatr(matr):
    det=matr[0]*matr[3]-matr[1]*matr[2]
    newMatr=[0,0,0,0]
    newMatr[0]=matr[3]/det
    newMatr[3]=matr[0]/det
    newMatr[1]=-matr[1]/det
    newMatr[2]=-matr[2]/det
    return newMatr




xArray = []
yArray = []
y1Array = []

p = -1
while p<1:
    xArray.append(p)
    y1Array.append( (5-(p**4))**(1/4) )
    yArray.append( (2.7)**(-p) )

    p+=0.0001
    
plt.plot (xArray,yArray)
plt.plot (xArray,y1Array)

plt.show()





i=0
matrix=[0,0,0,0]
x = -0.1
y = 1.1


methodNewtonaBool = True

while methodNewtonaBool:
    matrix[0]=dxf(x,y)
    matrix[1]=dyf(x,y)
    matrix[2]=dxg(x,y)
    matrix[3]=dyg()
    obrM = obratnaiaMatr(matrix)
    dx=-obrM[0]*f(x,y)-obrM[1]*g(x,y)
    dy=-obrM[2]*f(x,y)-obrM[3]*g(x,y)
    x=x+dx
    y=y+dy
    i+=1

    if (abs(dx)<e) and (abs(dy)<e):        
        print("Zadanie 3")    
        print("f(x,y): "+str(fZadanie3(x,y)))
        print("g(x,y): "+str(gRightZadanie3(x)))
        print("x: "+str(x)+" y: "+str(y))
        print("method Newton a ITERATIONS:",i)
        methodNewtonaBool = False
        

```
